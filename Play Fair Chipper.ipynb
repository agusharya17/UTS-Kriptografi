```python
def generate_playfair_matrix(key):
    # Membuat matriks Playfair dari kunci
    key = key.replace("J", "I")
    key = key.upper()

    # Membuat matriks alfabet
    alphabet = "ABCDEFGHIKLMNOPQRSTUVWXYZ"

    # Mengisi matriks dengan kunci
    key_matrix = []
    for char in key:
        if char not in key_matrix:
            key_matrix.append(char)

    # Mengisi matriks dengan alfabet yang tidak ada di kunci
    for char in alphabet:
        if char not in key_matrix:
            key_matrix.append(char)

    # Mengubah matriks ke dalam bentuk 5x5
    key_matrix = [key_matrix[i:i+5] for i in range(0, 25, 5)]

    return key_matrix

def find_char_positions(matrix, char):
    # Mencari posisi karakter dalam matriks Playfair
    for i in range(5):
        for j in range(5):
            if matrix[i][j] == char:
                return i, j

def encrypt_playfair(plaintext, key_matrix):
    # Enkripsi teks dengan Playfair Cipher
    ciphertext = ""
    plaintext = plaintext.replace("J", "I")
    plaintext = plaintext.upper()

    for i in range(0, len(plaintext), 2):
        char1, char2 = plaintext[i], plaintext[i+1]

        row1, col1 = find_char_positions(key_matrix, char1)
        row2, col2 = find_char_positions(key_matrix, char2)

        if row1 == row2:
            ciphertext += key_matrix[row1][(col1 + 1) % 5]
            ciphertext += key_matrix[row2][(col2 + 1) % 5]
        elif col1 == col2:
            ciphertext += key_matrix[(row1 + 1) % 5][col1]
            ciphertext += key_matrix[(row2 + 1) % 5][col2]
        else:
            ciphertext += key_matrix[row1][col2]
            ciphertext += key_matrix[row2][col1]

    return ciphertext

def decrypt_playfair(ciphertext, key_matrix):
    # Dekripsi teks yang telah dienkripsi dengan Playfair Cipher
    plaintext = ""
    ciphertext = ciphertext.upper()

    for i in range(0, len(ciphertext), 2):
        char1, char2 = ciphertext[i], ciphertext[i+1]

        row1, col1 = find_char_positions(key_matrix, char1)
        row2, col2 = find_char_positions(key_matrix, char2)

        if row1 == row2:
            plaintext += key_matrix[row1][(col1 - 1) % 5]
            plaintext += key_matrix[row2][(col2 - 1) % 5]
        elif col1 == col2:
            plaintext += key_matrix[(row1 - 1) % 5][col1]
            plaintext += key_matrix[(row2 - 1) % 5][col2]
        else:
            plaintext += key_matrix[row1][col2]
            plaintext += key_matrix[row2][col1]

    return plaintext

# Input kunci dari pengguna
key = input("Masukkan kunci (tanpa huruf J, ubah J menjadi I): ")

# Membuat matriks Playfair dari kunci
key_matrix = generate_playfair_matrix(key)

# Input teks dari pengguna
plaintext = input("Masukkan teks yang akan dienkripsi: ")

# Enkripsi teks
ciphertext = encrypt_playfair(plaintext, key_matrix)
print("Teks yang dienkripsi:", ciphertext)

# Dekripsi teks
decrypted_text = decrypt_playfair(ciphertext, key_matrix)
print("Teks yang didekripsi:", decrypted_text)
```
